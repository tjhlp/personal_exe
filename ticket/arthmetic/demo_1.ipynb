```python
import csv
```

```python
ticket_info = []
with open('./Ticket_Test.csv', 'r', encoding='utf-8') as f:
    # 通过文件对象创建 csv 写入对象
    content = csv.reader(f)
    # 把数据写入到 csv 文件中
    # 写入标题
    for row in content:
        # 去除无效数据
        if row:
            ticket_info.append(row)
    # 去除标题
    ticket_info.pop(0)
```

```python
info_ = ticket_info[:]
```

```python
info_[:-5]
```

```python
r_list = []
b_list = []
for temp_info in info_:
    temp_list = []
    r_ball_list = temp_info[1].split(' ')
    r_ball_list = [int(x) for x in r_ball_list]
    b_ball = int(temp_info[2][:2]) if len(temp_info[2]) > 2 else int(temp_info[2])
    b_list.append(b_ball)
    r_list.append(r_ball_list)
```

```python
r_list[:5]
```

```python
import numpy as np
import pandas as pd
```

```python
# 以线性回归进行预测
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler  # 标准化工具
from sklearn.metrics import mean_squared_error
```

```python
# 1.数据集划分
x_train, x_test, y_train, y_test = train_test_split(r_list, b_list, random_state=22)

# 2.特征工程-标准化
transfer = StandardScaler()
x_train = transfer.fit_transform(x_train)
x_test = transfer.fit_transform(x_test)

# 3.机器学习-线性回归(正规方程)
estimator = LinearRegression()
estimator.fit(x_train, y_train)

# 4.模型评估
# 4.1 获取系数等值
y_predict = estimator.predict(x_test)
print("预测值为:\n", y_predict)
print("模型中的系数为:\n", estimator.coef_)
print("模型中的偏置为:\n", estimator.intercept_)

# 4.2 评价
# 均方误差
error = mean_squared_error(y_test, y_predict)
print("误差为:\n", error)
```
